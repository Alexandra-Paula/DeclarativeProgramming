### Exercise 2: Convert Seconds to hh:mm:ss
Generate a random number between 0 and 10000 representing seconds.  
Calculate and display its representation in hours, minutes, and seconds (hh:mm:ss).  
Optionally, use the `datetime` module.

```python
import random 
import time

num = random.randrange(0, 10000)
print("Seconds:", num)

representation = time.strftime('%H:%M:%S', time.gmtime(num))
print("Seconds in (hh:mm:ss):", representation)
```

### Exercise 3: Lambda Functions for Arithmetic Operations
Define two float numbers and calculate their sum and difference using lambda functions.  
Apply all functions in a list on different types of data using the `map()` function.

```python
a = 13.3
b = 15.56

sum_func = lambda a, b : a + b
diff_func = lambda a, b : a - b
print(sum_func(a, b))
print(diff_func(a, b))

functions = [sum_func, diff_func]

listValues = [(100, 24),
              (101.12, 24.5)]

for func in functions:
    results = list(map(lambda pair: func(*pair), listValues))
    print(f"Result of {func.__name__ if hasattr(func, '__name__') else 'lambda'}: {results}")
```

### Exercise 4: Filter Distinct Values from a List
Define a list of integers and display only the distinct values using the `filter()` function.

```python
# filter() method
integersList = [1, 2, 2, 3, 4, 4, 5, 5, 6, 6, 6, 7, 7]
seenSet = set()
distinct_numbers = list(filter(lambda x: x not in seenSet and not seenSet.add(x), integersList))
print("Valori distincte:", distinct_numbers)

```

### Exercise 5: Count Character Occurrences using Reduce
Define a function that returns the number of occurrences of a character in a string.  
Use the `reduce()` function to implement this.

```python
from functools import reduce

def numberOccurrences(string, char) :
    return reduce(lambda acc, c: acc + (c == char), string, 0)


string = "Python"
character = "n"
print(f"The number of occurrences of '{character}' in '{string}' is : ", numberOccurrences(string, character))
```

```python
#count() method
string = "Python"
cnt = string.count('n')
print(cnt)
```


### Exercise 7: Lambda Functions for String Manipulation
Define a list of lambda functions that return:  
- The second character of a string  
- The string in uppercase  
- The position of a given character in the string  
Call each function in the list sequentially.

```python
lambdaFunctions = [
    lambda string, char=None: string[1],
    lambda string, char=None: string.upper(),
    lambda string, char=None: string.find(char)
]

str_val = "Python"
findChar = "y"

for func in lambdaFunctions:
    print(func(str_val, findChar))
```

### Exercise 8: Average of Three Grades with Default Values
Define a function that calculates the average of three grades.  
If not all grades are provided, use default values equal to 4.  
Call the function with different combinations of positional and keyword arguments.

```python
def averageGrades (grade1=4, grade2=4, grade3=4):
    avg = (grade1 + grade2 + grade3) / 3
    return round(avg, 2)

print(averageGrades(10, 10, 9))
print(averageGrades(9, 9))
```

### Exercise 9: Recursive Sum of First N Natural Numbers
Define a recursive function that returns the sum of the first N natural numbers.

```python
def recurSum(n):
    if n == 0:
        return 0
    return n + recurSum(n - 1)

if __name__ == "__main__":
    n = 3
    print(recurSum(n))
```
