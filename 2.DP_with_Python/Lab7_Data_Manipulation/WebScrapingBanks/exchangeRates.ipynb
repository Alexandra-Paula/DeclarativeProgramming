scraping-ul cursului valutar, banca micb

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

driver = webdriver.Chrome()
driver.get("https://micb.md/")

input("Press Enter when you have selected the date...")

csv_file = "exchange_rates_micb.csv"
header = ["Cumparare", "Vanzare", "Curs BNM", "Valuta", "Date"]
with open(csv_file, "w", newline="") as file:
    csv.writer(file).writerow(header)

valid_currencies = ["USD", "EUR", "RUB", "RON", "UAH", "GBP", "CHF"]

for day in range(1, 32):
    try:
        WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.table div.row")))

        rows = driver.find_elements(By.CSS_SELECTOR, "div.table div.row")
        data_for_day = []
        currency_index = 0

        for row in rows:
            cols = row.find_elements(By.CSS_SELECTOR, "div.col")
            raw_data = [col.text.strip() for col in cols]

            if len(cols) >= 3 and any(raw_data):
                buy, sell, bnm = (raw_data[0] if raw_data[0] != "-" else "-",
                                  raw_data[1] if raw_data[1] != "-" else "-",
                                  raw_data[2] if raw_data[2] != "-" else "-")
                data_for_day.append([buy, sell, bnm, valid_currencies[currency_index], f"2025-11-{day:02d}"])
                currency_index += 1
                if currency_index == len(valid_currencies):
                    break

        if data_for_day:
            with open(csv_file, "a", newline="") as file:
                csv.writer(file).writerows(data_for_day)
            print(f"Data for 2025-11-{day:02d} collected.")

    except Exception as e:
        print(f"Error processing data for 2025-11-{day:02d}: {e}")

    time.sleep(7)

driver.quit()
print(f"Data saved to {csv_file}.")
```

scraping-ul cursului valutar, victoriabank

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

driver = webdriver.Chrome()
driver.get("https://www.victoriabank.md/curs-valutar")

input("Press Enter when you have selected the date...")

csv_file = "exchange_rates_victoriabank.csv"
header = ["Cumparare", "Vanzare", "Curs BNM", "Valuta", "Date"]
with open(csv_file, "w", newline="") as file:
    csv.writer(file).writerow(header)

valid_currencies = ["USD", "EUR", "RUB", "RON", "UAH", "GBP", "CHF"]

for day in range(1, 32):
    try:
        WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.content-box.bg-white.js-option-2 .tr")))

        rows = driver.find_elements(By.CSS_SELECTOR, "div.content-box.bg-white.js-option-2 .tr")
        data_for_day = []
        found_currencies = []

        for row in rows:
            cols = row.find_elements(By.CSS_SELECTOR, "div.td")
            raw_data = [col.text.strip() for col in cols]

            if len(cols) >= 4 and any(raw_data):
                buy, sell, bnm = (raw_data[1] if raw_data[1] != "-" else "-", 
                                  raw_data[2] if raw_data[2] != "-" else "-", 
                                  raw_data[3] if raw_data[3] != "-" else "-")
                if raw_data[0] in valid_currencies:
                    data_for_day.append([buy, sell, bnm, raw_data[0], f"2025-11-{day:02d}"])
                    found_currencies.append(raw_data[0])

        for currency in valid_currencies:
            if currency not in found_currencies:
                data_for_day.append(["-", "-", "-", currency, f"2025-11-{day:02d}"])

        if data_for_day:
            with open(csv_file, "a", newline="") as file:
                csv.writer(file).writerows(data_for_day)
            print(f"Data for 2025-11-{day:02d} collected.")

    except Exception as e:
        print(f"Error processing data for 2025-11-{day:02d}: {e}")

    time.sleep(8)

driver.quit()
print(f"Data saved to {csv_file}.")

```
