1. Write the code sequence that groups the data from the three CSV files (obtained in the previous lab) based on one fundamental characteristic (for example: the exchange rate from three banks for the same day).

```python
import pandas as pd

file1 = "exchange_rates_victoriabank.csv"
file2 = "exchange_rates_micb.csv"
file3 = "exchange_rates_maib.csv"

df_victoriabank = pd.read_csv(file1)
df_micb = pd.read_csv(file2)
df_maib = pd.read_csv(file3)

df_victoriabank["Banca"] = "Victoriabank"
df_micb["Banca"] = "MICB"
df_maib["Banca"] = "MAIB"

df_victoriabank["Date"] = pd.to_datetime(df_victoriabank["Date"], errors="coerce")
df_micb["Date"] = pd.to_datetime(df_micb["Date"], errors="coerce")
df_maib["Date"] = pd.to_datetime(df_maib["Date"], errors="coerce")

combined_df = pd.concat([df_victoriabank, df_micb, df_maib])

grupate_df = combined_df[combined_df["Valuta"].isin(["EUR", "USD", "RON"])]

grupate_df['Cumparare'] = pd.to_numeric(grupate_df['Cumparare'].replace('-', None), errors='coerce')
grupate_df['Vanzare'] = pd.to_numeric(grupate_df['Vanzare'].replace('-', None), errors='coerce')

grupate_df["Cumparare_Column"] = grupate_df["Valuta"] + " Cump " + grupate_df["Banca"]
grupate_df["Vanzare_Column"] = grupate_df["Valuta"] + " Vanz " + grupate_df["Banca"]

cumparare_pivot = grupate_df.pivot_table(
    index="Date",
    columns="Cumparare_Column",
    values="Cumparare",
)

vanzare_pivot = grupate_df.pivot_table(
    index="Date",
    columns="Vanzare_Column",
    values="Vanzare",
)

final_pivot = pd.concat([cumparare_pivot, vanzare_pivot], axis=1).reset_index()

final_pivot.to_csv("Ex1_group_data.csv", index=False)

print("The regrouped data has been saved in 'Ex1_group_data.csv'.")
```

2. Determine the minimum, maximum, and average for the grouped values for each individual day, as well as the minimum, maximum, and average for all stored data.

```python
import pandas as pd

final_pivot = pd.read_csv("Ex1_group_data.csv")

cump_columns = [col for col in final_pivot.columns if 'Cump' in col]
vanz_columns = [col for col in final_pivot.columns if 'Vanz' in col]

daily_stats = pd.DataFrame()
daily_stats['Date'] = final_pivot['Date']

for col in cump_columns:
    moneda = col.split()[0]
    banci = [f"{moneda} Cump MAIB", f"{moneda} Cump MICB", f"{moneda} Cump Victoriabank"]

    daily_stats[f'{moneda} Cump MED'] = final_pivot[banci].mean(axis=1)
    daily_stats[f'{moneda} Cump Max'] = final_pivot[banci].max(axis=1)
    daily_stats[f'{moneda} Cump Min'] = final_pivot[banci].min(axis=1)

for col in vanz_columns:
    moneda = col.split()[0]
    banci = [f"{moneda} Vanz MAIB", f"{moneda} Vanz MICB", f"{moneda} Vanz Victoriabank"]

    daily_stats[f'{moneda} Vanz MED'] = final_pivot[banci].mean(axis=1)
    daily_stats[f'{moneda} Vanz Max'] = final_pivot[banci].max(axis=1)
    daily_stats[f'{moneda} Vanz Min'] = final_pivot[banci].min(axis=1)

daily_stats.to_csv("Ex2.group_data_per_day.csv", index=False)

overall_stats = pd.DataFrame()

for moneda in ['EUR', 'RON', 'USD']:
    cump_banci = [f"{moneda} Cump MAIB", f"{moneda} Cump MICB", f"{moneda} Cump Victoriabank"]
    overall_stats[f'{moneda} Cump MED'] = [final_pivot[cump_banci].mean(axis=0).mean()]
    overall_stats[f'{moneda} Cump Max'] = [final_pivot[cump_banci].max(axis=0).max()]
    overall_stats[f'{moneda} Cump Min'] = [final_pivot[cump_banci].min(axis=0).min()]

    vanz_banci = [f"{moneda} Vanz MAIB", f"{moneda} Vanz MICB", f"{moneda} Vanz Victoriabank"]
    overall_stats[f'{moneda} Vanz MED'] = [final_pivot[vanz_banci].mean(axis=0).mean()]
    overall_stats[f'{moneda} Vanz Max'] = [final_pivot[vanz_banci].max(axis=0).max()]
    overall_stats[f'{moneda} Vanz Min'] = [final_pivot[vanz_banci].min(axis=0).min()]

overall_stats.to_csv("Ex2.group_data_summary.csv", index=False, header=True)

print("The data with statistics for each day has been saved to 'Ex2.group_data_per_day.csv'.")
print("The summarized data for all days has been saved to 'Ex2.group_data_summary.csv'.")
```

3. Write the code sequence that will split the CSV file obtained after concatenation into N files, where each individual file will correspond to one week (7 rows). The files should be named after the first and last date they contain (e.g., 20231103_20231109.csv).

```python
import pandas as pd

data = pd.read_csv("Ex1_group_data.csv")
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

data = data.sort_values(by='Date')

num_rows = len(data)
weeks = num_rows // 7

for i in range(weeks):
    week_data = data.iloc[i*7:(i+1)*7]
    start_date = week_data['Date'].iloc[0].strftime('%Y%m%d')
    end_date = week_data['Date'].iloc[-1].strftime('%Y%m%d')

    file_name = f"{start_date}_{end_date}.csv"

    week_data.to_csv(file_name, index=False)
    print(f"Fișierul {file_name} a fost salvat.")

if num_rows % 7 != 0:
    week_data = data.iloc[weeks*7:]
    start_date = week_data['Date'].iloc[0].strftime('%Y%m%d')
    end_date = week_data['Date'].iloc[-1].strftime('%Y%m%d')

    file_name = f"{start_date}_{end_date}.csv"

    week_data.to_csv(file_name, index=False)
print(f"The file {file_name} was saved.")
```


4. Write a function that takes a datetime input and returns the rows from the DataFrame (from the concatenated data file) corresponding to that date, or None if no data exists for that date.

```python
import pandas as pd

data = pd.read_csv("Ex1_group_data.csv")
data['Date'] = pd.to_datetime(data['Date'], errors="coerce")

def get_data(input_date):
    if not isinstance(input_date, pd.Timestamp):
        raise ValueError("Data de intrare trebuie să fie de tip datetime")

    result = data[data['Date'] == input_date]

    if not result.empty:
        return "\n".join(
            " | ".join([f"{col}: {row[col]}" for col in result.columns])
            for _, row in result.iterrows()
        )
    else:
        return None

input_date = pd.Timestamp('2025-11-01')
result = get_data(input_date)
if result is not None:
    print(result)
else:
    print("Nu există date pentru această dată.")
```

5. Write the code sequence that allows displaying data for time intervals. For example, for each week separately (the function takes a datetime input) and returns the rows from the DataFrame for that period, or None if no data exists for the specified days.

```python
import pandas as pd

data = pd.read_csv("Ex1_group_data.csv")
data['Date'] = pd.to_datetime(data['Date'], errors="coerce")

def get_week_data(input_date):
    if not isinstance(input_date, pd.Timestamp):
        raise ValueError("Data de intrare trebuie să fie de tip datetime")

    week_start = input_date - pd.Timedelta(days=input_date.weekday())
    week_end = week_start + pd.Timedelta(days=6)

    result = data[(data['Date'] >= week_start) & (data['Date'] <= week_end)]

    if not result.empty:
        formatted_output = ""
        for idx, row in result.iterrows():
            formatted_output += " | ".join([f"{col}: {row[col]}" for col in result.columns]) + "\n"
        return formatted_output.strip()
    else:
        return None

input_date = pd.Timestamp('2025-11-03')
result = get_week_data(input_date)

if result is not None:
    print(result)
else:
    print("No data for this week.")
```
